**7. List Comprehension**

A short way to create lists in one line.

*Normal way*

```python
squares = []
for i in range(1,6):
    squares.append(i*i)
```

```python
squares
```

**List comprehension way**

```python
squares=[i*i for i in range(1,6)]
```

```python
squares
```
